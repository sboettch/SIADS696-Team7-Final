# Harmonization and Data Cleaning

This notebook documents how the clean repository combines CDC PLACES, EPA AirData, ACS, NOAA climate data, and Census county geography into a county-year modeling panel.


## Source Inputs

- CDC PLACES yearly files: `data/raw/cdc_places/`
- EPA AirData annual and daily AQI summaries: `data/raw/epa_airdata/`
- ACS county enrichment: `data/raw/acs/`
- NOAA county climate variables: `data/raw/noaa/`
- Census county boundaries: `data/raw/census_geography/`


## Harmonization Keys

The final panel is harmonized at county-year granularity. The preferred join key is five-digit county GEOID. County and state names are retained for validation, interpretation, and reporting, but they are not treated as substantive model predictors.


```python
from pathlib import Path
import pandas as pd

ROOT = Path.cwd()
panel = pd.read_csv(ROOT / 'data' / 'processed' / 'county_panel_enriched.csv', dtype={'geoid': str})
print(panel.shape)
display(panel.head())
```

## Cleaning and Imputation Steps

1. Aggregate PLACES files into a consistent county-year table.
2. Aggregate annual EPA AirData AQI variables and daily AQI exposure summaries.
3. Join ACS, NOAA, and Census geography enrichment by GEOID/year.
4. Engineer rates and ratios so counties are compared on intensity rather than raw size.
5. Use geographic imputation for sparse environmental fields where appropriate.
6. Build the supervised target by shifting FMD forward one year within county.
7. Exclude FMD, lower CI, and upper CI from unsupervised clustering features.
